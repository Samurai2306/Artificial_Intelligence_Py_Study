4. пример - График функции с заливкой

задача: Для функции sin(x·e^cos(x)) на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади используя функцию trapz().


```python
# 4.2.2
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

# 0 до 10 с шагом 0.05
x = np.arange(0, 10, 0.05)

#(x * e^cos(x))
y = np.sin(x * np.exp(np.cos(x)))

# методом трапеций
area = trapz(y, x)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label=r'$\sin(x \cdot e^{\cos(x)})$')
plt.fill_between(x, y, alpha=0.3, color='lightblue')
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title(r'График функции $\sin(x \cdot e^{\cos(x)})$ с заливкой', fontsize=14)
plt.legend(fontsize=11)
plt.axhline(y=0, color='k', linewidth=0.5)

#
plt.text(5, max(y)*0.8, f'Площадь = {area:.4f}', 
         fontsize=12, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

print(f"Площадь под графиком функции: {area:.4f}")
```

4. пример - Круговая и точечная диаграммы

```python
# 4.2.3
import matplotlib.pyplot as plt
import numpy as np

# Данные: оценки и количество студентов
grades = ['Отлично (5)', 'Хорошо (4)', 'Удовл. (3)', 'Неудовл. (2)']
counts = [8, 12, 6, 2]
colors = ['gold', 'lightgreen', 'lightblue', 'lightcoral']

# подграфиками
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Круговая диаграмма
ax1.pie(counts, labels=grades, colors=colors, autopct='%1.1f%%', 
        startangle=90, explode=(0.1, 0, 0, 0))
ax1.set_title('Круговая диаграмма распределения оценок', fontsize=14)

# Точечная диаграмма (scatter plot)
x_positions = np.arange(len(grades))
ax2.scatter(x_positions, counts, s=300, c=colors, alpha=0.6, edgecolors='black', linewidth=2)
ax2.set_xlabel('Оценки', fontsize=12)
ax2.set_ylabel('Количество студентов', fontsize=12)
ax2.set_title('Точечная диаграмма распределения оценок', fontsize=14)
ax2.set_xticks(x_positions)
ax2.set_xticklabels(grades)
ax2.grid(True, alpha=0.3)

#
for i, count in enumerate(counts):
    ax2.text(i, count + 0.3, str(count), ha='center', fontsize=11)

plt.tight_layout()
plt.show()

print("Статистика оценок:")
print(f"Всего студентов: {sum(counts)}")
for grade, count in zip(grades, counts):
    print(f"{grade}: {count} чел. ({count/sum(counts)*100:.1f}%)")
```

4. задание - График функции со срезом

```python
# 4.3.2
import numpy as np
import matplotlib.pyplot as plt

# Дана функция: (1/2) * cos(x) + sin(ln(x^2 + 3)) / e^x
# x изменяется от 1 до 10
x = np.linspace(1, 10, 10)

# Вычисляем значения функции
y = (1/2) * np.cos(x) + np.sin(np.log(x**2 + 3)) / np.exp(x)

# массива
x_slice = x[:5]
y_slice = y[:5]

print(f"Полный массив x: {x}")
print(f"Полный массив y: {y}")
print(f"\nСрез первой половины x: {x_slice}")
print(f"Срез первой половины y: {y_slice}")

# Построение графиков
plt.figure(figsize=(14, 6))

# массива
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', linewidth=2, marker='o', markersize=8, label='Полный массив')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Линейный график полного массива', fontsize=13)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

#
plt.subplot(1, 2, 2)
plt.scatter(x_slice, y_slice, s=150, c='red', alpha=0.6, edgecolors='black', linewidth=2, label='Срез (первая половина)')
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Точечный график среза', fontsize=13)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

#
for i, (xi, yi) in enumerate(zip(x_slice, y_slice)):
    plt.text(xi, yi + 0.02, f'({xi:.1f}, {yi:.3f})', fontsize=8, ha='center')

plt.tight_layout()
plt.show()
```

4. задание - Площадь под графиком

Задача: Дана функция |cos(x \* e^(cos(x) + ln(x+1)))|. Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним используя функцию trapz().


```python
# 4.3.3
import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

# 0.1 до 10 с шагом 1 (избегаем 0 из-за ln)
x = np.arange(0.1, 10.1, 1)

# Вычисляем значения функции |cos(x * e^(cos(x) + ln(x+1)))|
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

# методом трапеций
area = trapz(y, x)

# Построение графика
plt.figure(figsize=(12, 7))
plt.plot(x, y, 'b-o', linewidth=2, markersize=8, label=r'$|\cos(x \cdot e^{\cos(x) + \ln(x+1)})|$')
plt.fill_between(x, y, alpha=0.3, color='lightblue')
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=13)
plt.ylabel('y', fontsize=13)
plt.title(r'График функции $|\cos(x \cdot e^{\cos(x) + \ln(x+1)})|$ с заливкой', fontsize=14)
plt.legend(fontsize=12)
plt.axhline(y=0, color='k', linewidth=0.5)

#
plt.text(5, max(y)*0.8, f'Площадь под графиком:\n{area:.4f}', 
         fontsize=13, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7),
         ha='center')

plt.tight_layout()
plt.show()

print(f"Площадь под графиком функции: {area:.4f}")
print(f"\nТочки графика:")
for xi, yi in zip(x, y):
    print(f"  x = {xi:5.1f}, y = {yi:.6f}")
```

5. задание - Создание калькулятора

Задача: Создайте простейший калькулятор, включающий основные действия для двух переменных: +, -, \*, /, а также вычисление следующих функций: e^(x+y), sin(x+y), cos(x+y), x^y.


```python
# 4.3.5
import math

def calculator(x, y, operation):
    """
    Простейший калькулятор для двух переменных
    
    Параметры:
    x, y - числа
    operation - строка с кодом операции
    
    Доступные операции:
    '+', '-', '*', '/' - арифметические
    'exp' - e^(x+y)
    'sin' - sin(x+y)
    'cos' - cos(x+y)
    'pow' - x^y
    """
    
    if operation == '+':
        return x + y
    elif operation == '-':
        return x - y
    elif operation == '*':
        return x * y
    elif operation == '/':
        if y == 0:
            return "Ошибка: деление на ноль!"
        return x / y
    elif operation == 'exp':
        return math.exp(x + y)
    elif operation == 'sin':
        return math.sin(x + y)
    elif operation == 'cos':
        return math.cos(x + y)
    elif operation == 'pow':
        return x  y
    else:
        return "Неизвестная операция!"

# Тестирование калькулятора
x = 5
y = 3

print("---")
print(f"     КАЛЬКУЛЯТОР")
print("---")
print(f"x = {x}")
print(f"y = {y}")
print("---")

operations = [
    ('+', 'Сложение'),
    ('-', 'Вычитание'),
    ('*', 'Умножение'),
    ('/', 'Деление'),
    ('exp', 'Экспонента e^(x+y)'),
    ('sin', 'Синус sin(x+y)'),
    ('cos', 'Косинус cos(x+y)'),
    ('pow', 'Степень x^y')
]

for op, name in operations:
    result = calculator(x, y, op)
    print(f"{name:25} ({op:4}): {result}")

print("---")

# Интерактивная версия (раскомментируй для использования но если что ты не доделал оно как то через раз работает)
# print("\nИнтерактивный режим:")
# try:
#     x = float(input("Введите x: "))
#     y = float(input("Введите y: "))
#     print("\nДоступные операции: +, -, *, /, exp, sin, cos, pow")
#     operation = input("Введите операцию: ")
#     result = calculator(x, y, operation)
#     print(f"\nРезультат: {result}")
# except ValueError:
#     print("Ошибка: введите числа!")
```
